### Add column with conditional values


This code won't work:
```
if insurance_df['age'] > 62:
  insurance_df['retired'] = "yes"
else:
  insurance_df['retired'] = "no"
```
But this will:
```
df.loc[df['column name'] condition, 'new column name'] = 'value if condition is met'
```



````
# add a new column with condition
insurance_df.loc[insurance_df['age']>= 62, 'retired'] = "yes"
print(insurance_df[insurance_df['age']>= 62].head(3))
print(insurance_df.info())
````

Note that we have only **68 non-null** values for in `retired` column. That's because we didn't specify values for other cases.\

###Group by
By **group by** we are referring to a process involving one or more of the following steps:
*  **Splitting** the data into groups based on some criteria.
*  **Applying** a function to each group independently.
*  **Combining** the results into a data structure.

Out of these, the split step is the most straightforward. In fact, in many situations we may wish to split the data set into groups and do something with those groups.

Think of `groupby` as a pivot table with information inside based on function you choose: `.count()`, `.sum()`, `.mean()` etc.

````
# number of charges grouped by retired category
insurance_df[['region', 'charges']].groupby('region').count()
````

````
# mean values for all numerical columns grouped by retired category
insurance_df.groupby('region').mean()
````

````
# create a DataFrame with mean values of charges grouped by retired category
mean_by_region = insurance_df[['region','charges']].groupby('region').mean()
print(mean_by_region)

# plot a bar chart
mean_by_region.plot.barh()
````

````
# create a DataFrame with number of charges grouped by retired category
sum_by_region = insurance_df[['region','charges']].groupby('region').sum()
print(sum_by_region)

# plot a pie chart
sum_by_region.plot.pie(subplots=True,autopct='%.2f')
````

##BONUS: More about Vizuals

So far we used plots that are available in pandas, however there are numerous libraries that allow you to build nice plots with a few lines of code. Seaborn is one of them and you can look up some additional plots [here](https://seaborn.pydata.org/examples/index.html).

The relationship between **x** and **y** can be shown for different subsets of the data using the `hue`, `size`, and `style` parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e. both hue and style for the same variable) can be helpful for making graphics more accessible.
